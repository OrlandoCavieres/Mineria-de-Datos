- **Clasificador Multiclase:** Una tarea de clasificación con más de dos clases; Por ejemplo, clasificar un conjunto de imágenes de animales que pueden ser perros, gatos o tortugas. La clasificación de múltiples clases supone que cada muestra se asigna a una y solo una etiqueta: un animal puede ser una tortuga o un gato, pero no ambos al mismo tiempo.

### ***Breve descripción de los siguientes clasificadores, enfocado en como se utilizan en clasificación multiclase.***
  - **Naive Bayes:** Es uno de los algoritmos más simples y poderosos para clasificar, aunque sólo se basa en estimaciones. Se trata de una técnica de clasificación y predicción supervisada, la cual predice la probabilidad de posibles resultados. Asume que los atributos de la clase son independientes. Este clasificador está basado en el teorema de Bayes
  - **Support Vector Machine (SVM):** Se utilizan tanto para regresión como para clasificación. Es una técnica que encuentra la mejor separación posible entre clases, donde el SVM encuentra el hiperplano que maximiza el margen de separación entre clases.
  - **Perceptron Multicapa:** Es una red neuronal artificial (RNA) formada por múltiples capas, de tal manera que tiene capacidad para resolver problemas que no son linealmente separables, lo cual es la principal limitación del perceptrón (también llamado perceptrón simple). El perceptrón multicapa puede estar totalmente o localmente conectado. En el primer caso cada salida de una neurona de la capa "i" es entrada de todas las neuronas de la capa "i+1", mientras que en el segundo cada neurona de la capa "i" es entrada de una serie de neuronas (región) de la capa "i+1".

  >>  - **Capa de entrada, salida y oculta:** La capa de entrada está constituida por aquellas neuronas que introducen los patrones de entrada en la red, en estas neuronas no se produce procesamiento. La capa oculta está formada  por aquellas neuronas cuyas entradas provienen de capas anteriores y cuyas salidas pasan a neuronas de capas posteriores. La capa de salida se define como neuronas cuyos valores de salida se corresponden con las salidas de toda la red.  
  >>  - **Pesos y bias:** Los pesos son coeficientes que pueden adaptarse dentro de la red que determinan la intensidad de la señal de entrada registrada por la neurona artificial. Ellos son la medida de la fuerza de una conexión de entrada. Las bías controlan qué tan predispuesta está la neurona a disparar un 1 o un 0 independiente de los pesos. Un sesgo alto hace que la neurona requiera una entrada más alta para generar una salida de 1.

  >>  - **Función de activación:** Una vez realizada la suma ponderada la función de activación evalúa que tan aptas son las señales de entrada, si la sumatoria de estas rebasa un cierto umbral son mandadas como salida, de lo contrario se regresa al paso 2 y se ajustan los pesos (pueden ser generados de manera aleatoria).

  >> - **Funciones Sigmoid, Relu:** Una función sigmoide es una función matemática que tiene una curva característica en forma de "S" o curva sigmoide. Un ejemplo común de una función sigmoid es la función logística. La función sigmoide es continuamente diferenciable. Por el otro lado, las ReLUs permiten el paso de todos los valores positivos sin cambiarlos, pero asigna todos los valores negativos a 0. Aunque existen funciones de activación aún más recientes, la mayoría de las redes neuronales de hoy utilizan ReLU o una de sus variantes.

- **Explique la codificación one hot (one hot encoding), explique su utilidad:**  One-hot es un grupo de bits entre los cuales las combinaciones legales de valores son solo aquellas con un solo bit alto (1) y todas las demás bajas (0). Una implementación similar en la que todos los bits son '1', excepto uno '0', a veces se llama one-cold. En estadística, las variables dummyt representan una técnica similar para representar datos categóricos. La codificación One-hot se usa a menudo para indicar el estado de una máquina de estados . Cuando se utiliza código binario o gris, se necesita un decodificador para determinar el estado. Sin embargo, una máquina de estado one-hot, no necesita un decodificador como la máquina de estado está en el n-ésimo estado si y sólo si el n-ésimo bit es alto (1).

Explique en que consiste, su utilidad y describa diferentes estrategias:**  Es el proceso de seleccionar las más importante y/o relevante características de un conjunto de datos, con el objetivo de mejorar el rendimiento de predicción de los predictores, proporcionar predictores más rápidos y más rentables y proporcionar una mejor comprensión del proceso subyacente que generó los datos. Como estrategia se usan los metodos de filtro y envoltura, los metodos de filtro se utilizan generalmente como un paso de preprocesamiento de datos, la selección de características es independiente de cualquier algoritmo de Machine Learning. Los metodos de envoltura necesitan un algoritmo de Machine Learning y utilizan su rendimiento como criterio de evaluación. Este método busca una característica que sea más adecuada para el algoritmo y tiene como objetivo mejorar el rendimiento. Las características se clasifican según los puntajes estadísticos que tienden a determinar la correlación de las características con la variable de resultado, teniendo en cuenta que la correlación es un término muy contextual y varía de un trabajo a otro.

## Parte 2.2 Entrenamiento de Naive Bayes

Entrene un clasificador Naive Bayes para clasificación de las clases. Calcule métricas adecuadas sobre el conjunto de entrenamiento y test.

Busque una manera de calcular curva ROC para clasificación multiclase.

<h6 style="color: black; background-color:#FF9933; padding: 5px; border-radius: 5px; width: 95%; font-size: 16px">
    Función para obtener métricas que requieren binarización
</h6>

```python
def multiclass_metrics(real, predicted, average = 'macro'):
    lb = LabelBinarizer()
    lb.fit(real)
    real = lb.transform(real)
    predicted = lb.transform(predicted)
    rocauc = roc_auc_score(real, predicted, average = average)
    prec = precision_score(real, predicted, average = average)
    recall = recall_score(real, predicted, average = average)
    return rocauc, prec, recall
```

    Establecer sets de entrenamiento y de test para x e y
</h6>

```python
x_train = df_data_train.drop(columns = [48])
x_train
```

```python
y_train = df_data_train[48]
y_train
```

```python
x_test = df_data_test.drop(columns = [48])
x_test
```

```python
y_test = df_data_test[48]
y_test
```

<h6 style="color: black; background-color:#FF9933; padding: 5px; border-radius: 5px; width: 95%; font-size: 16px">
    Entrenamiento multiclase con Naive Bayes
</h6>

```python
gnb = GaussianNB()

bayes_time_i = datetime.now()

gnb.fit(x_train, y_train)
y_pred = gnb.predict(x_train)
y_pred_test = gnb.predict(x_test)
y_score_test = gnb.predict_proba(x_test)

bayes_time_completed = datetime.now() - bayes_time_i
```

Tanto precisión, como recall y area bajo la curva son calculados mediante estandarización binaria de los indicadores reales y predichos para obtener un valor promedio final que es usado finalmente como score para dicho atributo estadístico. Por lo tanto, miden el comportamiento globalizado de dichas métricas entre todas las clases.

F1 Score mide la precisión y la calidad del entrenamiento en el subset de test.

Se mide Accuracy para ambos subset para comprobar que no existe mucha diferencia entre ambos y validar el entrenamiento mediante el clasificador.

Se calcula el tiempo para establecer eficiencia del algoritmo.

```python
def create_styler(dataframe):
    dataframe_styled = dataframe.style\
        .set_properties(**{'background-color': '#E3E3E3', 'color': 'black', 'font-style': 'italic', 'font-size': '80%'})\
        .set_table_styles([dict(selector='td', props=[('text-align', 'center'), ('font-weight', 'bold')]),
                           dict(selector='th', props=[('background-color', '#51639E'), ('font-size', '80%'), 
                                                      ('font-weight', 'bold'), ('color', 'white')]),
                           dict(selector='td:hover', props=[('background-color', '#FBBCBC')])])
    return dataframe_styled
```

```python
indexes_summaries = ['Precisión', 'Exactitud Entrenamiento', 'Exactitud Test',
                     'Recall', 'F1 Score', 'Área Bajo la Curva', 'Tiempo de Entrenamiento']

roc_auc_nb, precision_nb, recall_nb = multiclass_metrics(y_test, y_pred_test)

df_nb_summary = pd.DataFrame(columns = ['Naive Bayes'], index = indexes_summaries)
df_nb_summary.at['Precisión', 'Naive Bayes'] = round(precision_nb, 5)
df_nb_summary.at['Exactitud Entrenamiento', 'Naive Bayes'] = round(accuracy_score(y_train, y_pred), 5)
df_nb_summary.at['Exactitud Test', 'Naive Bayes'] = round(accuracy_score(y_test, y_pred_test), 5)
df_nb_summary.at['Recall', 'Naive Bayes'] = round(recall_nb, 5)
df_nb_summary.at['F1 Score', 'Naive Bayes'] = round(f1_score(y_test, y_pred_test, average = 'weighted'), 5)
df_nb_summary.at['Área Bajo la Curva', 'Naive Bayes'] = round(roc_auc_nb, 5)
df_nb_summary.at['Tiempo de Entrenamiento', 'Naive Bayes'] = bayes_time_completed
df_nb_summary_styled = create_styler(df_nb_summary)
df_nb_summary_styled
```

```python
def plot_multiclass_confusion_matrix(conf_matrix):
    cm_mc = ff.create_annotated_heatmap(
        z = conf_matrix[::-1],
        x = [f'c{i+1}' for i in range(0, len(conf_matrix[0]))],
        y = [f'c{i+1}' for i in range(len(conf_matrix[0]) - 1, -1, -1)],
        colorscale = 'agsunset',
        xgap = 0.2,
        ygap = 0.2,
        showscale = True,
    )
    cm_mc.update_layout(
        width = 750,
        height = 550,
        xaxis_title = 'Predecido',
        yaxis_title = 'Real'
    )
    cm_mc.show()
```

```python
def create_metrics_dataframe(metrics_dict, name_model):
    df = pd.DataFrame(columns=[name_model], 
                      index=['Exactitud', 'Promedio Diagonal', 'Desviación Estándar Diagonal', 
                             'N° Épocas', 'Tiempo Entrenamiento'])
    df.at['Exactitud', name_model] = metrics_dict['exactitud']
    df.at['Promedio Diagonal', name_model] = metrics_dict['promedio_diagonal']
    df.at['Desviación Estándar Diagonal', name_model] = metrics_dict['desv_est_diagonal']
    df.at['N° Épocas', name_model] = metrics_dict['epocas']
    df.at['Tiempo Entrenamiento', name_model] = metrics_dict['tiempo']
    df_styled = create_styler(df)
    return df, df_styled
```

```python
def plot_loss(clasificador):
    pl_loss = go.Figure()
    pl_loss.add_trace(go.Scatter(
        y = clasificador.loss_curve_,
        marker = dict(color = '#6633FF')
    ))
    pl_loss.update_layout(
        plot_bgcolor = '#DCD8EC',
        width = 750,
        height = 550,
        xaxis_title = 'N° Época',
        yaxis_title = 'Función Loss',
        xaxis_tick0 = 0,
        xaxis_dtick = 50
    )
    pl_loss.show()
```

```python
def train_perceptron(train_x, train_y_onehot, test_x, test_y_onehot,
                     activator, hidden_layers, iterations = 1000, earlystop = False):

    time_i = datetime.now()

    mlp = MLPClassifier(activation=activator, hidden_layer_sizes=hidden_layers, max_iter=iterations, early_stopping=earlystop)
    mlp.fit(train_x, train_y_onehot)
    y_predicted_mlp = mlp.predict(test_x)

    time_training = datetime.now() - time_i

    mconf = confusion_matrix(test_y_onehot.argmax(axis=1), y_predicted_mlp.argmax(axis=1))
    class_report = classification_report(test_y_onehot, y_predicted_mlp)
    prom_diag = np.mean(np.diag(mconf))
    desv_est_diag = np.std(np.diag(mconf))
    exact = accuracy_score(test_y_onehot, y_predicted_mlp)
    total_iters = mlp.n_iter_
    metrics = {
        'reporte_clasificacion': class_report,
        'matriz_confusion': mconf,
        'promedio_diagonal': prom_diag,
        'desv_est_diagonal': desv_est_diag,
        'exactitud': exact,
        'epocas': total_iters,
        'tiempo': time_training
    }
    return mlp, metrics
```

<h5 style="color:white; background-color:#2E7EFF; text-align:center; padding: 5px; border-radius: 5px; font-size: 24px">
    Añadir secuencialmente características a un clasificador (Perceptron)
</h5>

El modelo elegido es el número 11 de ReLU con dos capas ocultas de 150 y 75 neuronas.

Se establecen a continuación una función para realizar la selección dentro del tipo de perceptron escogido y una función auxiliar para observar más fácilmente las características escogidas de las columnas del dataset.

```python
def train_mlp_selected_with_sfs(number_features, train_x, train_y_onehot, direction='forward'):
    time_i = datetime.now()

    mlp = MLPClassifier(activation='relu', hidden_layer_sizes=[150, 75], max_iter=1000, early_stopping=True)
    sfs = SequentialFeatureSelector(mlp, n_features_to_select=number_features, n_jobs=-1, direction=direction)
    sfs.fit(train_x, train_y_onehot)

    time_selecting = datetime.now() - time_i

    return sfs, time_selecting
```

```python
def print_atributes_supported(list_atributes):
    print('Atributos o columnas seleccionadas y soportadas:\n')
    str_list = [str(element) for element in list_atributes]
    print(f"{'  '.join(str_list)}\n")
```

Se establece el uso de **early_stopping = True** debido a que al procedimiento demoraba muchisimo en un inicio, ya que debe realizar combinaciones de todos los elementos posibles en múltiples iteraciones, tanto como elementos deben ser elegidos o desechados. Con esta opción se pasó de obtener un resultado en horas a solo minutos. Early Stopping permite obtener el punto mayor donde la validación y el entrenamiento alcanzan su máxima exactitud y menor valor de la función pérdida sin sobreentrenar para la prueba dada.

Se inicia las primeras selecciones más cercanas al cero usando una dirección fordward, ya que es más eficiente y más conveniente para términos de tiempo y obtención de resultados, ya que significa emplear menos iteraciones para llegar a ellos.

Se establece el uso de todos los procesadores de un computador para realizar los cálculos *(n_jobs = -1)*, lo que agiliza el trabajo al emplear procesadores en paralelo. El computador empleado tiene 8 núcleos o CPU, y 16 hilos, lo que aumenta significativamente el rendimiento del algoritmo.
